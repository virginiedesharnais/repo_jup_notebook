## 4) Extraire l’information concernant l’année de plantation pour l’ensemble des lignes
list_annee_plantation contient l'année de plantation pour tout le fichier 

```python
def recuperation_data(list_res, id_ligne, id_ligne_bis=0):
    new_dict = {}
    cpt = 0
    for ligne in list_res:
        if id_ligne_bis != 0:
            new_dict[ligne[id_ligne_bis]]=ligne[id_ligne]
        else :
            new_dict[cpt]=ligne[id_ligne]
        cpt = cpt +1
    return new_dict


list_annee_plantation = list(recuperation_data(list_res, 19, 0).values())
list_annee_plantation
```

Objectifs respectés pour les genres => ('Acer', 5276, 16.89%) < 20%
Objectifs non respectés pour les espèces => ('Platanus - acerifolia', 4422, 14.15%) > 10%

```python
def objectif(obj, pourcent, nom):
    txt = ""
    if float(pourcent) > float(obj) : 
        txt = "Objectifs de " + str(obj) + "% non respectés => "+ nom + " - " + str(pourcent) + "% > " + str(obj) +"%"
    else :
        txt ="Objectifs de " + str(obj) + "% respectés => " + nom + " - " + str(pourcent) + "% < " + str(obj) +"%"
    return txt
```

## 9) Diversifie-t-on plus maintenant qu’avant ? 

```python
#TENTATIVE 1 : 
#Création d'une liste affichant :
#les années, les genres, le nb d'arbres, le % par rapport au nb total d'arbres sur une tranche de dates spécifique
#Tentative foirée...

#Pour lundi : idée de Florian 
#=> sortir pour chaque année, le TOP 3 des genres et des espèces en % avec une 4ème part "Autres"

def get_list_bota_tranche(date, liste_bota_all_annee):
    set_bota_annee = set(liste_bota_all_annee)
    set_bota_annee.discard('')
    list_t_bota_annee_tranche = list()

    total_tranche = 0
    for t in set_bota_annee:
        if t[0] != "" and int(t[0]) < date:
            total_tranche = total_tranche + liste_bota_all_annee.count(t)

    for t in set_bota_annee : 
        if t[0] != "" and int(t[0]) < date:
            nb_arbres_annee = liste_bota_all_annee.count(t)
            list_t_bota_annee_tranche.append((t[0], t[1],nb_arbres_annee, round((nb_arbres_annee/total_tranche)*100, 2)) )
    list_t_bota_annee_tranche.sort(reverse=True, key=lambda t:t[1])
    return list_t_bota_annee_tranche

get_list_bota_tranche(1970, list_bota_annee)
```
